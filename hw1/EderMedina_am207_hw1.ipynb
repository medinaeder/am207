## Problem 2: Maximum Likelihood Estimates

**Coding required for Part B only**

Suppose you observe the following data set $\mathbf{x}^{(0)} = (0.5, 2.5), \mathbf{x}^{(1)} = (3.2, 1.3), \mathbf{x}^{(2)} = (2.72, 5.84), \mathbf{x}^{(3)}= (10.047, 0.354)$. By convention, for any vector $\mathbf{x}$, we will denote the first component of $\mathbf{x}$ by $x_{1}$ and the second component by $x_{2}$. Suppose that the data is drawn from the same two-dimensional probability distribution with pdf $f_X$, that is, $\mathbf{x}^{(i)} \overset{iid}{\sim} f_X$, where
$$
f_X(\mathbf{x}) =  4\lambda_1^2 x_{1}x_{2} \mathrm{exp} \left\{-\lambda_0 (x^2_{1} + x^2_{2}) \right\}.
$$
You should assume that $\lambda_1, \lambda_0 > 0$ and that $f_X$ is supported on the nonnegative quandrant of $\mathbb{R}^2$ (i.e. $f_X$ is zero when either component is negative). 

### Part A:

What are the values for $\lambda_0$ and $\lambda_1$ that maximizes the likelihood of the observed data? 

Answer:
In order to compute the values of $\lambda_0$ and $\lambda_1$ that maximizes the likelihood of the observed data we must start of with a normalized PDF. 
That is we want

\begin{equation}
\int_0^{\infty} \int_0^{\infty} f_X(x_1, x_2) dx_1 dx_2 =  1
\end{equation}

\begin{equation}
4 \lambda_1^2 \int_0^{\infty} x_1 \mathrm{exp}(\lambda_0 x_1^2)\int_0^{\infty} x_2 \mathrm{exp}(\lambda_0 x_2^2)\:dx_2 dx_1
=1 
\end{equation}
where by substituting $u = \lambda_0 x_i^2$ first for $i = 2$ then for $i = 1$, which leads to the relation
\begin{equation}
\frac{4 \lambda_1^2}{4\lambda_0^2} = 1
\end{equation}

In order for our pdf to be normalized, it is required for $\lambda_1 = \lambda_0$. Our pdf is then simplified described by

\begin{equation}
f_X(x_1,x_2)= 4 \lambda_0^2 x_{1} x_{2} \mathrm{exp} \left \{-\lambda_0 (x^2_{1} + x^2+_2) \right\} 
\end{equation}

Next, we determine the value for $\lambda$ that is a maximum likelihood estimate. The maximum likelihood is determined by the product of the probabilities for each data set

\begin{equation}
L(\lambda_0) = \prod \limits_i^4 4 \lambda_0^2 x_{i1} x_{i2} \mathrm{exp} \left\{-\lambda_0 (x^2_{i1}+x^2_{i2}) \right\}
\end{equation}

It is easier to however compute the log likelihood given by

\begin{equation}
\ell(\lambda_0) = 4\ln(\lambda_0^2)+\sum \limits_i^4 \left\{\ln (4 x_{i1} x_{i2}) -\lambda_0 (x^2_{i1}+x^2_{i2})\right\}
\end{equation}

Differentiate with respect to $\lambda_0$ to find the extremum.

\begin{equation}
\frac{\partial \ell}{\partial \lambda_0} = \frac{8}{\lambda_0} -\sum \limits_i^4 (x^2_{i1}+x^2_{i2}) = 0 \implies \lambda_0 = \frac{8}{\sum \limits_i^4 (x^2_{i1}+x^2_{i2})}
\end{equation}

### Empirically Determine the mean of $f_{x_1}$ ###

In order to determine the mean of $f_{x_1}$, we need to resample the data set. 

```python
M_samples = 10000
N_points  = df.x1.size
bs_np = np.random.choice(df.x1, size = (M_samples, N_points), replace = True)
sd_mean=np.mean(bs_np, axis=1)
sd_std=np.std(bs_np, axis=1)
plt.hist(sd_mean, bins=30, normed=True, alpha=0.5,label="samples");
plt.axvline(df.x1.mean(), 0, 1, color='r', label='Our Sample')
plt.legend()

fx1_mean = np.mean(sd_mean)
fx1_se = np.std(sd_mean)

print("Empirical Mean: ", fx1_mean)
print("Emperical Standard Error: ",fx1_se)
```

### Empirically Determine the standard deviation of $f_{x_2|x_1}$ for $x_1 \in (3.99, 4.01)$

```python
data = df.x2[(df.x1 >3.99) & (df.x1 < 4.01)]
N_points  = data.size
bs_np = np.random.choice(data, size = (M_samples, N_points), replace = True)
sd_mean=np.mean(bs_np, axis=1)
sd_std=np.std(bs_np, axis=1)
plt.hist(sd_mean, bins=30, normed=True, alpha=0.5,label="samples");
plt.axvline(data.mean(), 0, 1, color='r', label='Our Sample')
plt.legend()

fx1_mean = np.mean(sd_std)
fx1_se = np.std(sd_std)
print("Standard Error", fx1_mean)
print("Standard Error of Standard error",fx1_se)
```
